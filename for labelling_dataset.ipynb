```python
import os
import pandas as pd
import json

# Mount Google Drive if using Google Colab
from google.colab import drive
drive.mount('/content/drive')

# Path to the folder containing videos
videos_folder = '/content/drive/My Drive/dfdc_train_part48_preprocessed'

# Path to the globalMetaData.json
global_metadata_path = '/content/drive/MyDrive/ML dataset/metadata_dfdc48.json'

# Path to save the smallMeta.csv
small_metadata_path = '/content/drive/MyDrive/ML dataset/smallMeta12_04.csv'

# Get list of video files in the folder
video_files = os.listdir(videos_folder)
video_files = [file for file in video_files if file.endswith('.mp4')]

# Load JSON metadata
with open(global_metadata_path, 'r') as file:
    global_metadata = json.load(file)

# Create a list to store filename-label pairs
file_label_pairs = []

# Iterate through video files and find their labels
for video_file in video_files:
    if video_file in global_metadata:
        label = global_metadata[video_file]['label']
        file_label_pairs.append([video_file, label])

# Convert list to DataFrame
small_metadata = pd.DataFrame(file_label_pairs, columns=['filename', 'label'])

# Save small metadata to CSV
small_metadata.to_csv(small_metadata_path, index=False)

print("Small metadata CSV created successfully.")
```
